```python
import matplotlib.pyplot as plt
import scipy.optimize as opt
import numpy as np

timer = [0, 2, 3, 6, 7, 9, 12, 13, 15, 16, 19, 21, 22, 24]
høyde = [114, 94, 96, 140, 156, 167, 123, 103, 83, 88, 141, 164, 161, 131]

def f(x, a, b, c, d, e, f):
    return a*x**5 + b*x**4 + c*x**3 + d*x**2 + e*x + f

popt, pcov = opt.curve_fit(f, timer, høyde)
print(popt)

xverdier = np.linspace(0,30, 1000)
#print(xverdier)

yverdier = f(xverdier, popt[0], popt[1], popt[2], popt[3], popt[4], popt[5])

plt.plot(xverdier, yverdier)
plt.plot(timer, høyde, '.')
plt.show()
```

```python
def g(x, A, c, fi, d):
    return A*np.sin(c*x+fi) + d

popt2, pcov2 = opt.curve_fit(g, timer, høyde)
print(popt2)

yverdier2 = g(xverdier, popt2[0], popt2[1], popt2[2], popt2[3])

plt.plot(xverdier, yverdier2)
plt.plot(timer, høyde, '.')
plt.show()
```
