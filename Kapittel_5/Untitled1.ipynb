```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

sim = 100_000
terning = [1, 2, 3, 4, 5, 6]

for ant_terninger in range(2, 21, 2):
    kastene = []
    for n in range(sim):
        kast = np.random.choice(terning, ant_terninger)
        kastene.append(sum(kast))
    
    plt.title(f'{ant_terninger} terning(er)')
    
    utfall = np.arange(ant_terninger, ant_terninger*6+2)
    plt.hist(kastene, bins=utfall-0.5, edgecolor='White', density=True)
    
    x = np.linspace(ant_terninger, ant_terninger*6, 1000)
    mu = np.mean(kastene)
    sigma = np.std(kastene)
    plt.plot(x, norm.pdf(x, mu, sigma), label=f'μ: {mu:.1f}, σ: {sigma:.1f}')
    plt.legend()
    plt.show()
```
