```python
from random import choice
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

ant_sim = 100_000
terning = [1,2,3,4,5,6]

ant_terninger = 10

kast = [sum(np.random.choice(terning, ant_terninger)) for x in range(ant_sim)]

utfall = np.arange(ant_terninger, ant_terninger*6+2)

plt.hist(kast, bins=utfall-0.5, density=True, color='green', edgecolor='white')

μ = np.mean(kast)
σ = np.std(kast)

x = np.linspace(ant_terninger, ant_terninger*6, 1000)

plt.plot(x, norm.pdf(x, μ, σ), color='red')

plt.show()
```
