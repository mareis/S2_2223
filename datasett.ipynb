```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
```

```python
kvinner['Høyde'].plot.hist(bins=30, density=True)

μ = kvinner['Høyde'].mean()
σ = kvinner['Høyde'].std()
minn = kvinner['Høyde'].min()
maks = kvinner['Høyde'].max()

x = np.linspace(minn, maks, 1000)

plt.plot(x, norm.pdf(x, μ, σ), color='red')

plt.show()
```
